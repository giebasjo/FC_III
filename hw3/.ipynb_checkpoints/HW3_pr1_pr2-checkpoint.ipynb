```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import urllib.request
import html5lib
import lxml

from urllib.request import urlopen
from bs4 import BeautifulSoup
from mpl_toolkits.mplot3d import Axes3D
```

# Problem 1 
## Part (a)

```python
# Define string
url = "http://www.treasury.gov/resource-center/data-chart-center/interest-rates/Pages/TextView.aspx?data=yieldYear&year=2018"

daily_yield_curves = np.matrix(pd.read_html(url, attrs={'class':'t-chart'})[0]).tolist()
```

```python
fout = open('daily_yield_curves.txt', 'w')
for row in daily_yield_curves:
    fout.write(str(row) + "\n")
    
fout.close()
```

## Part (b)

```python
dyc_df = pd.DataFrame(daily_yield_curves, dtype=np.float64)
```

```python
Y = np.array(dyc_df.columns)
dyc_df.columns = pd.Index(np.array(dyc_df.loc[0,:]))
dyc_df = dyc_df.drop(0, axis=0)
```

```python
dyc_df['days_passed'] = np.array(dyc_df.index) - 1
```

```python
X = np.array(dyc_df.index) - 1
Y = np.array([float(12*i) for i in [1/12, 3/12, 6/12, 1, 2, 3, 5, 7, 10, 20, 30]])

X, Y = np.meshgrid(X,Y)
```

```python
Z = np.array([np.array(dyc_df.iloc[i,1:12], dtype=np.float64) for i in range(0,69)]).T
```

```python
fig = plt.figure(figsize=(12,8))
ax = fig.gca(projection='3d')
ax.plot_surface(X,Y,Z)
ax.set_xlabel("Days since Jan. 2, 2018")
ax.set_ylabel("Months to Maturity")
ax.set_zlabel("Yield")
ax.set_title("Surface Plot")
```

```python
fig = plt.figure(figsize=(12,8))
ax = fig.gca(projection='3d')
ax.plot_wireframe(X,Y,Z)
ax.set_xlabel("Days since Jan. 2, 2018")
ax.set_ylabel("Months to Maturity")
ax.set_zlabel("Yield")
ax.set_title("Surface Plot")
```

# Problem 2

For this problem, we ran each of the parts in the terminal and stroed the results in a file to keep track of these. Hence, the data presented below is just c/p from that file.

```python
arr_size = np.array([1000,2000,4000,8000,16000,32000,64000,128000])
qs_times = np.array([0.0026,0.0053,0.0122,0.0287,0.0743,0.1508,0.4226,0.4226])
is_times = np.array([0.0625,0.2307,0.9292,3.7807,15.0723,15.0723,256.9330,1069.7866])
ss_times = np.array([0.0439,0.1697,0.6988,2.8699,11.3957,46.2508,188.4959,768.2181565899401])
ms_times = np.array([0.0040,0.0091,0.0199,0.0452,0.0452,0.1939,0.4229,0.8882])
cs_times = np.array([0.0007,0.0013,0.0014,0.0020,0.0056,0.0087,0.0130,0.0231])
```

```python
# plot features
plt.figure(figsize=(12,8))
plt.xlabel("Array size")
plt.ylabel("log(Sorting Time) (s)")
plt.title("log(SortingTime) vs. Array Size")

# plot content
plt.plot(arr_size, np.log(qs_times))
plt.plot(arr_size, np.log(is_times))
plt.plot(arr_size, np.log(ss_times))
plt.plot(arr_size, np.log(ms_times))
plt.plot(arr_size, np.log(cs_times))

plt.legend(['Quick', 'Insertion', 'Selection', 'Merge', 'Counting'], loc='upper left')
```

Relative to the C++ performance for each of these algorithms, Python doesn't perform as well. The ordering in the performance presented above is still as suspected though. <br>
At the top, Insertion and Selection sort perform roughly the same, as they are $\Theta(n^2)$<br>
In the middle, Quick and Merge sort perform roughly the same, as they are $\Theta(nlog(n))$<br>
The lowest is Counting sort, which is $\Theta(n+k)$. $k$ is negligble in these cases, so this is essentially a linear algorithm.
